```python
#   -> cal
def cal (num1, num2, op):
    """  op      """
    # Docstring:     
    if(op=="-"):
        result = num1-num2
        return result
    elif(op=="+"):
        result = num1+num2
        return result
    else:
        return ""
```

```python
# num1 , num2, op 
num1 = int(input("    >>"))
num2 = int(input("    >>"))
op = input(" (+,-) >>")
#  
result = cal(num1 , num2 , op)
# 
print(f":{result}")
```

```python
def all_sum(*args): #arguments
    result = 0
    for i in args :
        result +=i
    return result
```

```python
all_sum(1,3,5,7,9)
```
