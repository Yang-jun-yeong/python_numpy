```python
# 두개의 정수를 더해주는 함수 정의
def num_sum (num1,num2):
    result = num1+num2
    return result
# 아직까지는 함수가 정의된것!
# 사용을 하기 위해서는 함수를 호출!
```

```python
# 함수 호출
num_sum(1,3)
```

```python
# 함수 정의 -> cal
def cal (num1, num2, op):
    """입력 받는 op에 따라서 덧셈 혹은 뺄셈을 수행하는 함수"""
    # Docstring: 함수에 대한 설명을 작성하는 공간
    if(op=="-"):
        result = num1-num2
        return result
    elif(op=="+"):
        result = num1+num2
        return result
    else:
        return "잘못입력하셨습니다"
```

```python
# num1 , num2, op 입력받기
num1 = int(input("첫 번째 정수를 입력 >>"))
num2 = int(input("두 번째 정수를 입력 >>"))
op = input("연산자 입력(+,-) >>")
# 함수 호출
result = cal(num1 , num2 , op)
# 출력
print(f"결과:{result}")
```

```python
def all_sum(*args): #arguments
    result = 0
    for i in args :
        result +=i
    return result
```

```python
all_sum(1,3,5,7,9)
```

### 함수의 기본값

```python
# 함수를 정의할 때 기본값을 지정할 수 있음
# 숫자데이터에 제곱을 계산해주는 함수 (지수연산자**)
def power_num(num1,power = 2): # power를 입력해주지 않으면 기본 2제곱
    return num1**power
```

```python
# 함수호출
power_num(3)
```

```python
# 속성을 변경 -> 속성명 = 값
power_num(3 , power = 3)
```
