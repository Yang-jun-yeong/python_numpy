```python
import pandas as pd
```

```python
#   (data2019, data2020, data2021)
# : ""    ,  encoding = "euc-kr"

data2019 = pd.read_csv("data/2019.csv", index_col = "", encoding = "euc-kr")
data2020 = pd.read_csv("data/2020.csv", index_col = "", encoding = "euc-kr")
data2021 = pd.read_csv("data/2021.csv", index_col = "", encoding = "euc-kr")
data2021
```

```python
# 2021  "" 
data2021.drop("" , inplace = True)
```

```python
#      -> ""  
data2019[""] = data2019.loc[:,"":""].sum(axis = 1)
data2019
```

```python
data2020[""] = data2020.loc[:,"":""].sum(axis = 1)

data2021[""] = data2021.loc[:,"":""].sum(axis = 1)
```

```python
#    ""     (temp2019, temp2020, temp2021) 
temp2019 = data2019[data2019[""] ==""]
temp2020 = data2020[data2020[""] ==""]
temp2021 = data2021[data2021[""] ==""]
temp2021
```

```python
#   ""   - > Series  (s2019,s2020,s2021) 
s2019 = temp2019[""]
s2020 = temp2020[""]
s2021 = temp2021[""]
```

```python
# Series    (?    !) 2019, 2020, 2021 
s2019.name = "2019"
s2020.name = "2020"
s2021.name = "2021"
s2021
```

```python
#  (2019-2020, 2020-2021)
ch19_20 = (s2020-s2019) / s2019 * 100
ch20_21 = (s2021-s2020) / s2020 * 100
#  (2019-2020 , 2020-2021 )
ch19_20.name = "2019-2020"
ch20_21.name = "2020-2021"
```

```python
#  ()
result = pd.concat([s2019, ch19_20, s2020, ch20_21, s2021],axis = 1)
result
```
