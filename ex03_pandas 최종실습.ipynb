```python
import pandas as pd
```

```python
# 데이터 불러오기 (data2019, data2020, data2021)
# 조건: "관서명" 컬럼을 인덱스로 설정 , 인코딩 encoding = "euc-kr"

data2019 = pd.read_csv("data/2019.csv", index_col = "관서명", encoding = "euc-kr")
data2020 = pd.read_csv("data/2020.csv", index_col = "관서명", encoding = "euc-kr")
data2021 = pd.read_csv("data/2021.csv", index_col = "관서명", encoding = "euc-kr")
data2021
```

```python
# 2021년도에만 존재하는 "광주지방경찰청" 삭제
data2021.drop("광주지방경찰청" , inplace = True)
```

```python
# 각 연도별 범죄현황 총합 구하기 -> "총합" 컬럼 추가
data2019["총합"] = data2019.loc[:,"살인":"폭력"].sum(axis = 1)
data2019
```

```python
data2020["총합"] = data2020.loc[:,"살인":"폭력"].sum(axis = 1)

data2021["총합"] = data2021.loc[:,"살인":"폭력"].sum(axis = 1)
```

```python
# 각 연도별로 구분이 "발생건수" 인 데이터만 추출해서 임시변수 (temp2019, temp2020, temp2021) 담기
temp2019 = data2019[data2019["구분"] =="발생건수"]
temp2020 = data2020[data2020["구분"] =="발생건수"]
temp2021 = data2021[data2021["구분"] =="발생건수"]
temp2021
```

```python
# 각 연도별로 "총합" 컬럼만 추출하기 - > Series 형태로 추출될것임(s2019,s2020,s2021) 담기
s2019 = temp2019["총합"]
s2020 = temp2020["총합"]
s2021 = temp2021["총합"]
```

```python
# Series 데이터 이름 변경 (왜? 병합할건데 이름이 다 총합이니까!) 2019총계, 2020총계, 2021 총계
s2019.name = "2019총계"
s2020.name = "2020총계"
s2021.name = "2021총계"
s2021
```

```python
# 증감률 구하기(2019-2020, 2020-2021)
ch19_20 = (s2020-s2019) / s2019 * 100
ch20_21 = (s2021-s2020) / s2020 * 100
# 이름 바꿔주기(2019-2020 증감률, 2020-2021 증감률)
ch19_20.name = "2019-2020증감률"
ch20_21.name = "2020-2021증감률"
```

```python
#데이터 병합하기 (순서확인)
result = pd.concat([s2019, ch19_20, s2020, ch20_21, s2021],axis = 1)
result
```
